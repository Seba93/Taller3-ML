**2.2 Resolución del problema mediante árbol de regresión**

Para realizar el entrenamiento del árbol, se excluye la variable *datetime*, ya que se considera que la fecha exacta de arriendo no aporta información al problema, como si lo pueden hacer las variables *holiday* y *workingday*. Además, se omiten del modelo las variables *casual* y *registered*, pues sus valores pueden estar condicionados a otras variables, es decir, se piensa que no resultarían buenos predictores, y como el objetivo es predecir la variable *count*, se eliminan, en consecuencia, del modelo. Así, se proceden separar las variables predictivas de la variable a predecir en cada conjunto de datos.

```python
Xdf_train = dftrain.ix[:,['season','holiday','workingday','weather','temp','atemp', 'humidity','windspeed','hour']]
Xdf_val = dfval.ix[:,['season','holiday','workingday','weather','temp','atemp', 'humidity','windspeed','hour']]
Xdf_test = dftest.ix[:,['season','holiday','workingday','weather','temp','atemp', 'humidity','windspeed','hour']]
Ydf_train = dftrain.ix[:,'count']
Ydf_val = dfval.ix[:,'count']
Ydf_test = dftest.ix[:,'count']

X_train = Xdf_train.values
X_val = Xdf_val.values
X_test = Xdf_test.values
Y_train = Ydf_train.values
Y_val = Ydf_val.values
Y_test = Ydf_test.values
```

Luego, se está en condiciones de construir el modelo. 

```python
model = Tree(random_state=0)
model.fit(X_train,Y_train)
```

A continuación, se muestra la precisión lograda por el modelo sobre el conjunto de prueba.

```python
score_test = model.score(X_test,Y_test)
print "Precisión sobre conjunto de prueba = %f"%score_test
```

Se procede a implementar la función *eval_bikemodel*, que será utilizado para computar la función de error definida en el enunciado para este acercamiento y los de las siguientes secciones.

```python
def eval_bikemodel(y_predict,y_true): 
    diff = np.log(y_true+1.0) - np.log(y_predict+1.0)
    return np.sqrt(np.sum(np.square(diff))/len(y_predict))
```

Así, se procede a predecir valores de la variable *count* en los tres conjuntos de datos, y luego, calcular el valor que toma la función implementada en cada caso.

```python
Y_pred_train = model.predict(X_train)
Y_pred_val = model.predict(X_val)
Y_pred_test = model.predict(X_test)

kagg_train = eval_bikemodel(Y_pred_train,Y_train)
kagg_val = eval_bikemodel(Y_pred_val,Y_val)
kagg_test = eval_bikemodel(Y_pred_test,Y_test)
```

A continuación, se muestra el valor que adquiere la función para cada conjunto de datos. 

```python
print "Error sobre conjunto de entrenamiento = %f"%kagg_train
print "Error sobre conjunto de prueba = %f"%kagg_test
print "Error sobre conjunto de validación = %f"%kagg_val
```

Se aprecia que el error sobre el conjunto de prueba alcanza un valor no menor. Por otro lado, el error sobre el conjunto de entrenamiento es muy pequeño. Estos antecedentes permiten afirmar que el modelo está sobreajustado y por tanto no es el más adecuado para evaluar casos nuevos.

En base a los resultados anteriores, se construyen gráficos para comparar, en cada caso, los valores reales de *count* junto con los valores predichos. Primero, se elabora gráfico para conjunto de entrenamiento.

```python
plt.plot(Y_train, Y_pred_train, '.')
plt.xlabel('Cantidad real de arriendos')
plt.ylabel('Cantidad predicha de arriendos')
plt.show()
```

El gráfico construido confirma la hipótesis de sobreajuste del modelo, pues los puntos forman una recta casi perfecta, lo que en otras palabras quiere decir que los valores son prácticamente idénticos entre si. 

Ahora, se construye gráfico análogo al anterior, pero para el conjunto de pruebas.

```python
plt.plot(Y_test,Y_pred_test,'.')
plt.xlabel('Cantidad real de arriendos')
plt.ylabel('Cantidad predicha de arriendos')
plt.show()
```

En constraste con el gráfico anterior, se aprecia una alta dispersión entre los datos, lo que quiere decir que no existe una relación adecuada entre los valores reales y los valores que han sido predecidos, siendo esta situación reflejo del error cercano a 0.57 que se ha obtenido para este conjunto.

Otro aspecto interesante de analizar, es la importancia asociada a cada atributo presente en el modelo. 

```python
labels = ['season','holiday','workingday','weather','temp','atemp', 'humidity','windspeed','hour']
importances = model.feature_importances_
for i in range (0, len(labels)):
    print labels[i], ':', importances[i]*100, '%'
```

De acuerdo a estos valores, es la hora del día (*hour*) la variable que mayor importancia tiene para la predicción de valores, con aproximadamente un 60% de peso, seguida muy de lejos por la temperatura (*temp*) y la humedad (*humidity*). 

**2.3 Optimización de árbol de regresión modificando su profundidad**

Se busca disminuir el error de cómputo obtenido en la sección anterior. Para ello, se modifican los hiperparámetros asociados a él. En primer lugar, se estudia el cambio que se produce en el modelo al limitar la profundidad máxima que puede tener el árbol, lo que puede ser útil, ya que, en teoría, para una profundidad mayor del árbol, el algoritmo entrega la posibilidad de aprender relaciones entre variables de forma más específica. Notar que la evaluación del modelo se realiza sobre el conjunto de validación. A continuación, se muestra como varía el error del modelo en base a la profundidad máxima posible del árbol.

```python
depths = np.arange(1, 36)
error_func = []
for depth in depths:
    model = Tree(random_state=0,max_depth=depth)
    model.fit(X_train,Y_train)
    Y_pred_val = model.predict(X_val)
    kagg_val = eval_bikemodel(Y_pred_val,Y_val)
    error_func.append(kagg_val)

plt.plot(depths,error_func,'-')
plt.xlabel('Profundidad maxima del arbol')
plt.ylabel('Error')
plt.show()
```

Se observa como, para valores pequeños de profundidad, se obtienen errores extremadamente grandes. Sin embargo, el error comienza a decrecer hasta alcanzar su valor mínimo para aproximadamente 10 niveles, profundidad con la que se consigue un error cercano 0.5, pero inferior a este valor. Luego, se observa que el error comienza a aumentar hasta mantenerse relativamente constante en el valor 0.55. El valor exacto que se obtiene para el error con profundidad máxima 10 es 0.48, disminuyendo considerablemente con respecto al error de 0.55 que se obtiene al no limitar la profundidad del árbol.

```python
print 'Error con profundidad 10:', error_func[9]
```

**2.4 Efecto de considerar día de la semana como variable**

Se estudia el efecto de considerar el día de la semana en que se produce un arriendo. Dicho día es determinado en base a la fecha en que se realizó el arriendo (es decir, a partir de la variable *datetime*). Luego, se procede a enumerar los días, siendo lunes = 0, martes = 1 y así sucesivamente. Se denotará la nueva variable con el nombre *cday*, para cada uno de los conjuntos de datos.

Al comparar los errores obtenidos con los de la sección anterior, se puede ver que el error de validación disminuye levemente (de 0.48 a 0.47). Además el error de prueba también disminuye. Por otro lado, el error de entrenamiento aumenta considerablemente. Luego, de acuerdo a estos antecedentes, se observa una reducción considerable en el sobreajuste del modelo. Si bien la nueva variable es sólo la cuarta más importante, se tomará en cuenta para evaluaciones posteriores, dada su influencia en la reducción de overfitting. 

**2.5 Resolución del problema mediante SVR no lineal**

Se construye un modelo SVR no lineal, omitiendo las mismas variables que en el caso del árbol de regresión. Igualmente, se vuelven a agregar las variables *hour* y *cday*. Posterior a ello, se preprocesan los datos. El primer paso de preprocesamiento consiste en convertir cada variable discreta (*season*, *weather*, *hour* y *cday*) a un vector binario. De esta forma, si se considera la variable *season*, por ejemplo, la casilla j del vector binario asociado al registro i toma el valor 1 si el dato i registra un arriendo para *season* = j, 0 en caso contrario. 
EL segundo y último paso de preprocesamiento consiste en normalizar los datos de las variables restantes, de tal manera que dichos datos siguen una distribución gaussiana de media 0 y varianza 1. Considerar que se aplicado la función logaritmo sobre los valores de la variable a predecir, con el fin de evitar la indefinición de la función de error.

```python
#Se cargan los datos
dftrain = pd.read_csv('bike_sharing_train.csv')
dfval = pd.read_csv('bike_sharing_val.csv')
dftest = pd.read_csv('bike_sharing_test.csv')

dftrain['hour'] = pd.to_datetime(dftrain['datetime']).apply(lambda x: int(x.strftime('%H')))
dfval['hour'] = pd.to_datetime(dfval['datetime']).apply(lambda x: int(x.strftime('%H')))
dftest['hour'] = pd.to_datetime(dftest['datetime']).apply(lambda x: int(x.strftime('%H')))

dftrain['cday'] = pd.to_datetime(dftrain['datetime']).dt.dayofweek
dftrain['cday'] = pd.to_numeric(dftrain['cday'])
dfval['cday'] = pd.to_datetime(dfval['datetime']).dt.dayofweek
dfval['cday'] = pd.to_numeric(dfval['cday'])
dftest['cday'] = pd.to_datetime(dftest['datetime']).dt.dayofweek
dftest['cday'] = pd.to_numeric(dftest['cday'])

Xdf_train = dftrain.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday']]
Xdf_val = dfval.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday']]
Xdf_test = dftest.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday']]
#Ydf_train = dftrain.ix[:,'count']
Ydf_train =np.log(dftrain.ix[:,'count'])
#Ydf_val = dfval.ix[:,'count']
Ydf_val = np.log(dfval.ix[:,'count'])
#Ydf_test = dftest.ix[:,'count']
Ydf_test = np.log(dftest.ix[:,'count'])

#Conversion de variables discretas a vectores binarios
Xdf_train = pd.get_dummies(Xdf_train,columns=['season', 'weather','hour','cday'])
Xdf_val = pd.get_dummies(Xdf_val,columns=['season', 'weather','hour','cday'])
Xdf_test = pd.get_dummies(Xdf_test,columns=['season', 'weather','hour','cday'])

Xdf_val.insert(13,'weather_4', 0.0)
Xdf_test.insert(13,'weather_4', 0.0)

X_train = Xdf_train.values
X_val = Xdf_val.values
X_test = Xdf_test.values
Y_train = Ydf_train.values
Y_val = Ydf_val.values
Y_test = Ydf_test.values

#Normalizacion de datos
scalerX = StandardScaler()
X_train = scalerX.fit_transform(X_train)
X_val = scalerX.fit_transform(X_val)
X_test = scalerX.transform(X_test)
```

Luego, se procede a construir modelo SVR no lineal, con parámetro de regularización C = 1.0 y utilización de la función *rbf* (*radial basis function*). 

```python
model = SVR()
model.fit(X_train,Y_train)
Y_pred_train = model.predict(X_train)
Y_pred_val = model.predict(X_val)
Y_pred_test = model.predict(X_test)
kagg_train = eval_bikemodel(Y_pred_train,Y_train)
kagg_val = eval_bikemodel(Y_pred_val,Y_val)
kagg_test = eval_bikemodel(Y_pred_test,Y_test)
```

Se muestra el valor que adquiere la función para cada conjunto de datos. 

```python
print "Error sobre conjunto de entrenamiento = %f"%kagg_train
print "Error sobre conjunto de prueba = %f"%kagg_test
print "Error sobre conjunto de validación = %f"%kagg_val
```

**2.6 Optimización de SVR no lineal al modificar hiper-parámetros**

Se estudia el efecto de variar el parámetro de regularización C con el fin de minimizar los errores obtenidos para cada conjunto de datos. Para ello, se consideraron los valores de C = [0.01, 0.1, 1, 10, 100]. Para mostrar la evolución del error de validación en función del valor de C, se confecciona el siguiente gráfico.

```python
Cs = [0.01, 0.1, 1, 10 , 100]
error_val = []
for c in Cs:
    model = SVR(C=c,epsilon=0.01)
    model.fit(X_train, Y_train)
    Y_pred_train = model.predict(X_train)
    Y_pred_val = model.predict(X_val)
    Y_pred_test = model.predict(X_test)
    kagg_val = eval_bikemodel(Y_pred_val,Y_val)
    error_val.append(kagg_val)

plt.plot(Cs,error_val,'-')
plt.xlabel('Parametro de regularizacion')
plt.ylabel('Error de validacion')
plt.show()
```

Luego, el gráfico permite ver que para el parámetro C = 1, el error de validación alcanza su valor mínimo, aumentando progresivamente para valores superiores de C. Para valores inferiores de C, dicho error es del orden de 0.15. 

**2.7 Uso de validación cruzada para la selección de hiperparámetros**

**2.7.1 Árbol de clasificación**

Se considera el modelo óptimo obtenido en la sección 2.4. Primero, se investiga el efecto de usar el conjunto de validación para el proceso de entrenamiento y viceversa.

```python
dftrain = pd.read_csv('bike_sharing_train.csv')
dfval = pd.read_csv('bike_sharing_val.csv')
dftest = pd.read_csv('bike_sharing_test.csv')

dftrain['hour'] = pd.to_datetime(dftrain['datetime']).apply(lambda x: int(x.strftime('%H')))
dfval['hour'] = pd.to_datetime(dfval['datetime']).apply(lambda x: int(x.strftime('%H')))
dftest['hour'] = pd.to_datetime(dftest['datetime']).apply(lambda x: int(x.strftime('%H')))

Xdf_train = dftrain.ix[:,['season','holiday','workingday','weather','temp','atemp', 'humidity','windspeed','hour']]
Xdf_val = dfval.ix[:,['season','holiday','workingday','weather','temp','atemp', 'humidity','windspeed','hour']]
Xdf_test = dftest.ix[:,['season','holiday','workingday','weather','temp','atemp', 'humidity','windspeed','hour']]
Ydf_train = dftrain.ix[:,'count']
Ydf_val = dfval.ix[:,'count']
Ydf_test = dftest.ix[:,'count']

X_train = Xdf_train.values
X_val = Xdf_val.values
X_test = Xdf_test.values
Y_train = Ydf_train.values
Y_val = Ydf_val.values
Y_test = Ydf_test.values

dftrain['cday'] = pd.to_datetime(dftrain['datetime']).dt.dayofweek
dftrain['cday'] = pd.to_numeric(dftrain['cday'])
dfval['cday'] = pd.to_datetime(dfval['datetime']).dt.dayofweek
dfval['cday'] = pd.to_numeric(dfval['cday'])
dftest['cday'] = pd.to_datetime(dftest['datetime']).dt.dayofweek
dftest['cday'] = pd.to_numeric(dftest['cday'])

Xdf_train = dftrain.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday']]
Xdf_val = dfval.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday']]
Xdf_test = dftest.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday']]

X_train = Xdf_train.values
X_val = Xdf_val.values
X_test = Xdf_test.values

model = Tree(random_state=0, max_depth=10)
model.fit(X_val,Y_val)
Y_pred_train = model.predict(X_val)
Y_pred_val = model.predict(X_train)
Y_pred_test = model.predict(X_test)
kagg_train = eval_bikemodel(Y_pred_train,Y_val)
kagg_val = eval_bikemodel(Y_pred_val,Y_train)
kagg_test = eval_bikemodel(Y_pred_test,Y_test)
```

Se muestra el error obtenido para cada conjunto de datos.

```python
print "Error sobre conjunto de entrenamiento = %f"%kagg_train
print "Error sobre conjunto de prueba = %f"%kagg_test
print "Error sobre conjunto de validación = %f"%kagg_val
```

Luego, no es una buena medida invertir roles entre el conjunto de entrenamiento y el conjunto de validación, ya que tanto el error de prueba como de validación aumentan considerablemente. Además, el modelo se sobreajusta, al verificar la disminución del error sobre el conjunto de entrenamiento.

A continuación, se estudia la estimación de los hiper parámetros del modelo mediante k-fold cross validation. Para k = 10, validación cruzada encuentra el error mínimo con profundidad = 9, un nivel menos que el obtenido al no usar esta técnica. Sin embargo, el error que se obtiene es considerablemente más alto. Para cambiar esta situación, se observó que, al aumentar progresivamente el valor de k y mantener la profundidad constante, el error disminuía más mientras mayor era el valor de K. En el siguiente código, se muestra el error para profundidad = 9 y k = 100, parámetros para los cuáles se obtiene un error casi idéntico al que se logra cuando no se utiliza validación cruzada.

```python
k_fold = cross_validation.KFold(len(X_train), 100)
score_cv = 0
for k, (train, val) in enumerate(k_fold):
    model = Tree(random_state=0, max_depth=9)
    model.fit(X_train[train], Y_train[train])
    Ypred_val = model.predict(X_train[val])
    Ytrue_val = Y_train[val]
    score_fold = eval_bikemodel(Ypred_val,Ytrue_val)
    score_cv += score_fold
score_cv = score_cv / 100
print "Error por medio de 100-fold cross validation:", score_cv
```

**2.7.2 SVR no lineal**

Se considera el modelo óptimo obtenido en la sección 2.6. Primero, se indaga el efecto de usar el conjunto de validación para el proceso de entrenamiento y viceversa. Al ejecutar esta pieza de código, se muestran los errores obtenidos para cada conjunto de datos.

```python
#Se cargan los datos
dftrain = pd.read_csv('bike_sharing_train.csv')
dfval = pd.read_csv('bike_sharing_val.csv')
dftest = pd.read_csv('bike_sharing_test.csv')

dftrain['hour'] = pd.to_datetime(dftrain['datetime']).apply(lambda x: int(x.strftime('%H')))
dfval['hour'] = pd.to_datetime(dfval['datetime']).apply(lambda x: int(x.strftime('%H')))
dftest['hour'] = pd.to_datetime(dftest['datetime']).apply(lambda x: int(x.strftime('%H')))

dftrain['cday'] = pd.to_datetime(dftrain['datetime']).dt.dayofweek
dftrain['cday'] = pd.to_numeric(dftrain['cday'])
dfval['cday'] = pd.to_datetime(dfval['datetime']).dt.dayofweek
dfval['cday'] = pd.to_numeric(dfval['cday'])
dftest['cday'] = pd.to_datetime(dftest['datetime']).dt.dayofweek
dftest['cday'] = pd.to_numeric(dftest['cday'])

Xdf_train = dftrain.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday']]
Xdf_val = dfval.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday']]
Xdf_test = dftest.ix[:,['season','holiday','workingday','weather','temp','atemp','humidity','windspeed','hour','cday']]
#Ydf_train = dftrain.ix[:,'count']
Ydf_train =np.log(dftrain.ix[:,'count'])
#Ydf_val = dfval.ix[:,'count']
Ydf_val = np.log(dfval.ix[:,'count'])
#Ydf_test = dftest.ix[:,'count']
Ydf_test = np.log(dftest.ix[:,'count'])

#Conversion de variables discretas a vectores binarios
Xdf_train = pd.get_dummies(Xdf_train,columns=['season', 'weather','hour','cday'])
Xdf_val = pd.get_dummies(Xdf_val,columns=['season', 'weather','hour','cday'])
Xdf_test = pd.get_dummies(Xdf_test,columns=['season', 'weather','hour','cday'])

Xdf_val.insert(13,'weather_4', 0.0)
Xdf_test.insert(13,'weather_4', 0.0)

X_train = Xdf_train.values
X_val = Xdf_val.values
X_test = Xdf_test.values
Y_train = Ydf_train.values
Y_val = Ydf_val.values
Y_test = Ydf_test.values

#Normalizacion de datos
scalerX = StandardScaler()
X_train = scalerX.fit_transform(X_train)
X_val = scalerX.fit_transform(X_val)
X_test = scalerX.transform(X_test)

model = SVR(C=10,epsilon=0.01)
model.fit(X_val, Y_val)
Y_pred_train = model.predict(X_val)
Y_pred_val = model.predict(X_train)
Y_pred_test = model.predict(X_test)
kagg_train = eval_bikemodel(Y_pred_train,Y_val)
kagg_val = eval_bikemodel(Y_pred_val,Y_train)
kagg_test = eval_bikemodel(Y_pred_test, Y_test)
print "Error de entrenamiento = %f"%kagg_train
print "Error de validación = %f"%kagg_val
print "Error de test = %f"%kagg_test
```

Al igual que en el caso de árbol de regresión, no es conveniente invertir las matrices, pues los errores de validación y test aumentan.

A continuación, se estudia la estimación de los hiper parámetros del modelo mediante k-fold cross validation. Aumentar el valor k implica a su vez incrementar el costo computacional y SVR es un método de por si caro. Por ello, estudiará sólo un caso: k = 10 y C = 10.

```python
k_fold = cross_validation.KFold(len(X_train),10)
score_cv = 0
for k, (train, val) in enumerate(k_fold):
    model = SVR(C=10,epsilon=0.01)
    model.fit(X_train[train], Y_train[train])
    Ypred_val = model.predict(X_train[val])
    Ytrue_val = Y_train[val]
    score_fold = eval_bikemodel(Ypred_val,Ytrue_val)
    score_cv += score_fold
score_cv = score_cv / 10
print "Error por medio de 10-fold cross validation:", score_cv
```

Se observa que el error aumenta en comparación a no utilizar k-fold cross validation.

**2.8 Uso de ensamblaje de dos máquinas de aprendizaje para resolución del problema**

En esta sección se ensamblan dos máquinas de aprendizaje para resolver el problema en cuestión. Una de ellas es usada para la predicción de la demanda casual (variable *casual*) y la otra para la predicción de la demanda registrada (variable *registered*). Al sumar los dos resultados anteriores, se determina la demanda total (variable *count*). Se utilizaran como máquinas dos árboles de regresión. Si bien en las secciones anteriores se comprobó que el error obtenido mediante una SVR no lineal es menor con respecto al uso de un árbol de regresión, se tiene que el primero de estos métodos es más caro computacionalmente. Es por esta razón que se llevará a cabo un ensamblaje de dos máquinas: Se espera encontrar un método eficiente y que al mismo obtenga un error bajo, aunque esta última idea no es más que una hipótesis por ahora y deberá ser confirmada o bien refutada. Se comienza importando y transformando los conjuntos de datos tal como se hizo en secciones anteriores, pero ahora se consideran tres vectores Y para cada conjunto de datos: Los asociados a *casual*, *registered* y *count*.

**2.9 Uso de Random Forest para predicción de demanda total**

En esta sección, se hace uso del algoritmo de ensamblaje Random Forest para la resolución del problema. Para ello, se estudia el efecto de aumentar la cantidad de árboles usados. Se crea un gráfico para mostrar la evolución del error de validación en función de dicha cantidad. La profundidad de cada árbol ha sido seteada en 10, como consecuencia de los reultados de las secciones anteriores.

```python
estimators = np.arange(1, 100)
error_val = []

for n in estimators:
    model = RandomForestRegressor(n_estimators=n,max_depth=10,random_state=0)
    model.fit(X_train, Y_train_count)
    Y_pred_val = model.predict(X_val)
    kgg_val = eval_bikemodel(Y_pred_val, Y_val_count)
    error_val.append(kgg_val)
    
plt.plot(estimators,error_val,'-')
plt.xlabel('Cantidad de arboles utilizados')
plt.ylabel('Error de validacion')
plt.show()
```

El gráfico confeccionado muestra como, al usar pocos árboles, el error es del orden de 0.5, pero se reduce drásticamente al utilizar alrededor una cantidad cercana a 10 árboles. El error continúa disminuyendo para cantidades superiores, aunque en una proporción menor. Este último antecedente es fundamental, considerando que a medida que aumenta la cantidad de árboles considerados, el error disminuye, pero el costo computacional aumenta cada vez.
